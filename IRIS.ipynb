```python
df1=df.copy()
df1.drop('Species',axis=1,inplace=True)
```

```python
fig, ax = plt.subplots(2, figsize=(10, 8))
ax[0].scatter(df['PetalLengthCm'] ,df['PetalWidthCm'],c= df['Species'].map(colors), cmap='cool')
ax[0].legend(['with spcies parameter'])
ax[1].scatter(df1['PetalLengthCm'] ,df1['PetalWidthCm'], cmap='cool')
ax[1].legend(['without spcies parameter'])
plt.show()
```

```python
X=df1.copy()
plt.figure(figsize=(20, 10))
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method  = "ward"))

plt.title('Dendrogram')
plt.xlabel('species')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean')
```

```python
y_hc=hc.fit_predict(X)
```
